```python
import pandas as pd
import glob

csv_files = glob.glob(r'data/downloads/*.csv')

vars=['T', 'PRENEI', 'PRELIQ', 'FF', 'Q', 'DLI', 'SSI', 'HU', 'EVAP', 'ETP', 'PE', 'SWI', 'SSWI_10J', 'DRAINC', 'RUNC', 'RESR_NEIGE', \
      'RESR_NEIGE6', 'HTEURNEIGE', 'HTEURNEIGE6', 'HTEURNEIGEX', 'SNOW_FRAC', 'ECOULEMENT', 'WG_RACINE', 'WGI_RACINE', 'TINF_H', 'TSUP_H']

for var in vars:
    dfs = [pd.read_csv(f, sep=';', usecols=['LAMBX', 'LAMBY', 'DATE', var]) for f in csv_files[:]]
    df_total = pd.concat(dfs, ignore_index=True)
    df_total['DATE'] = pd.to_datetime(df_total['DATE'], format='%Y%m%d')
    df_total = df_total.rename(columns={'LAMBX': 'x', 'LAMBY': 'y', 'DATE': 'time'})
    ds = df_total.set_index(['time', 'y', 'x']).to_xarray()
    ds.to_netcdf(f'data/raw_per_var/QUOT_SIM2_var_{var}.nc', format='NETCDF4')
```

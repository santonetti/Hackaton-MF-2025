```python
# Décompression des fichiers .gz dans le répertoire spécifié
import glob
import gzip
import shutil

gz_files = glob.glob(r'C:\DEVS\HACKATON\data\*.gz')

for gz_file in gz_files:
    output_file = gz_file[:-3]
    with gzip.open(gz_file, 'rb') as f_in:
        with open(output_file, 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)
```
